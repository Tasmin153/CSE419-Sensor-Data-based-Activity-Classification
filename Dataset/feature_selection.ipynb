# 1.13.1 Removing features with low variance

```python
from sklearn.feature_selection import VarianceThreshold

def remove_feature_one(data,threshold=0):
        
    temp = VarianceThreshold(threshold=threshold)
    temp.fit_transform(data)
    return data[data.columns[temp.get_support(indices=True)]]
```

```python
f_s11 = remove_feature_one(data=d1)    
f_s11.head()
# my_threshold = (.8 * (1 - .8))
# f_s12 = remove_feature_one(data=d1, threshold=my_threshold)
```

# 1.13.4.1. L1-based feature selection

```python
from sklearn.svm import LinearSVC
from sklearn.feature_selection import SelectFromModel


lsvc = LinearSVC(C=0.01, penalty="l1", dual=False).fit(d1, d1_label)
model = SelectFromModel(lsvc, prefit=True)
feature_idx = model.get_support()
feature_name = d1.columns[feature_idx]

X_new = model.transform(d1)
X_new.shape
```

# 1.13.4.2. Tree-based feature selection

```python
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.feature_selection import SelectFromModel

clf = ExtraTreesClassifier(n_estimators=50)
clf = clf.fit(d1, d1_label)
model = SelectFromModel(clf, prefit=True)

feature_idx = model.get_support()
feature_name = d1.columns[feature_idx]

X_new = model.transform(d1)
X_new.shape               
```

# 1.13.5. Feature selection as part of a pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectFromModel
clf = RandomForestClassifier()
clf = clf.fit(d1, d1_label)
model = SelectFromModel(clf, prefit=True)

feature_idx = model.get_support()
feature_name = d1.columns[feature_idx]

X_new = model.transform(d1)
X_new.shape
```
