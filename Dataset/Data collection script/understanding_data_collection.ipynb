```python
import pandas as pd
from glob import glob
```

```python
file_list= []

for fname in glob("../../casas-dataset/*/*ann.features.csv"):
    #print(fname)
    file_list.append(fname)

print("Total files: ",len(file_list))
file_list
```

```python
def readable(filename):
    try:
        df = pd.read_csv(filename)
        df.head(5)
    except:
        print("not readable")
```

```python
for i in file_list:
    readable(i)
    
```
